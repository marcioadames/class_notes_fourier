# Série de Fourier de funções de período $2\pi$

Podemos usar a célula abaixo para criar séries com coeficientes $a_n$ e $b_n$

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-2*np.pi,2*np.pi,500)

#y = np.heaviside(x,0)*(x-np.pi)**2/np.pi+np.heaviside(-x,0)*(x+np.pi)**2/np.pi

# constante da série
a0 = 1/2

def an(n):
    value = 1/n
    return value
   
def bn(n):
    value = 0
    return value

def f(x, N):
    fun=0*x+a0
    for i in range(1,N+1):
        fun = fun+an(i)*np.cos(i*x)+bn(i)*np.sin(i*x)
    return fun

y2 = f(x, 9)

#plt.plot(x, y)
plt.plot(x, y2)
plt.show()
```

Podemos usar a célula abaixo para plotar a série de fourier de uma função

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-np.pi,np.pi,500)

y = np.heaviside(x,0) -1/2

def an(n):
    c = y*np.cos(n*x)
    value = c.sum()/c.size
    if n==0:
        value=value
    else:
        value=value*2
    return value
   
def bn(n):
    c = y*np.sin(n*x)
    return c.sum()/c.size*2

def f(x, N):
    fun=0*x
    for i in range(N+1):
        fun = fun+an(i)*np.cos(i*x)+bn(i)*np.sin(i*x)
    return fun

y2 = f(x, 20)

plt.plot(x, y)+plt.plot(x, y2)
plt.show()
```
