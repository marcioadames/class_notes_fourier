Podemos usar a célula abaixo para criar séries com coeficientes $a_n$ e $b_n$. Use para testar os coeficientes que encontrou

```python
import numpy as np
import matplotlib.pyplot as plt

#semi-período
L = 1

#número de cosenos e senos na série 
num_termos = 20

x = np.linspace(-3*L,3*L,500)

#y = np.heaviside(x,0)*(x-np.pi)**2/np.pi+np.heaviside(-x,0)*(x+np.pi)**2/np.pi

# a0 inicial da série
a0 = 1/2

def an(n):
    value = 1/n
    return value
   
def bn(n):
    value = 0
    return value

def f(x, N):
    fun=0*x+a0
    for i in range(1,N+1):
        fun = fun+an(i)*np.cos(i*np.pi*x/L)+bn(i)*np.sin(i*np.pi*x/L)
    return fun

y2 = f(x, num_termos)

#plt.plot(x, y)
plt.plot(x, y2)
plt.show()
```

Podemos usar a célula abaixo para plotar a expansão par e a expansão iímpar de uma série

```python
import numpy as np
import matplotlib.pyplot as plt

#semi-período
L = 3

#número de cosenos e senos na série 
num_termos = 20


x = np.linspace(0,L,500)

y = 3*np.heaviside(x-1,0)-3*np.heaviside(x-L,0)

xx = np.linspace(-3*L,3*L,1500)

def an(n):
    c = y*np.cos(n*np.pi*x/L)
    value = c.sum()/c.size
    if n==0:
        value=value
    else:
        value=value*2
    return value
   
def bn(n):
    c = y*np.sin(n*np.pi*x/L)
    return c.sum()/c.size*2

def f(x, N):
    fun=0*x
    for i in range(N+1):
        fun = fun+an(i)*np.cos(i*np.pi*x/L)
    return fun


def g(x, N):
    fun=0*x
    for i in range(N+1):
        fun = fun+bn(i)*np.sin(i*np.pi*x/L)
    return fun


y2 = f(xx, num_termos)
plt.title('Expansão par')
plt.grid()
plt.ylim((-4,4))
plt.plot(xx, y2)+plt.plot(x, y)
plt.show()

y2 = g(xx, num_termos)
plt.title('Expansão ímpar')
plt.grid()
plt.ylim((-4,4))
plt.plot(xx, y2)+plt.plot(x, y)
plt.show()
```
